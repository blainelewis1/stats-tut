## Explore Figures

```python
aggregations = {
    'Participant' : {
        'x' : 'Condition',
        'kind' : 'bar',
        'hue' : 'Participant',
        'data' : training_block_grouped,
    },
    
    
#     'Trial' : {
#         'x' : 'Trial',
#         'hue' : 'Participant',
#         'row' : 'Condition',
#         'legend' : False,
#         'data' : training,
#     },
    
    'Block' : {
        'x' : 'Block',
        'hue' : 'Condition',
        'data' : training_block_grouped,
    },
    
    'Last Blocks' : {
        'x' : 'Grouped',
        'hue' : 'Condition',
        'kind' : 'bar',
        'data' : training_last_blocks_grouped,
    },
}
```

```python
tab = widgets.Tab()
tabs = []
for i , (aggregation, kwargs) in enumerate(aggregations.items()):
    accordion = widgets.Accordion()
    tabs.append(accordion)
    children = []
    tab.set_title(i, aggregation)

    for i, variable in enumerate(training_aggregations.keys()):
        accordion.set_title(i, variable)
        output = widgets.Output()
        children.append(output)
        with output:
            sns.factorplot(y=variable, 
                           **kwargs,
                           size=height,
                           aspect=width/height
                          )
            plt.show()

    accordion.children = children
tab.children = tabs
tab
```
